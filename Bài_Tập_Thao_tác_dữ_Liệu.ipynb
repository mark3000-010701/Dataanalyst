# 4. Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.

```python
df5 =df.groupby(['type_of_land'])['price'].mean().reset_index(name='Giá trung bình cộng')
df5
```

```python
df['Giá Trung bình cộng'] = df5['Giá trung bình cộng']
df
```

```python
df7 =df.groupby(['type_of_land'])['price'].max().reset_index(name='Giá lớn nhất')
df7
```

```python
df['giá max'] = df5['Giá lớn nhất']
df
```

```python
df6 =df.groupby(['type_of_land'])['price'].min().reset_index(name='Giá nhỏ nhất')
df6
```

```python
df5['Giá max'] = df7['Giá lớn nhất']
df5['Giá Min'] = df6['Giá nhỏ nhất']
df5
```
