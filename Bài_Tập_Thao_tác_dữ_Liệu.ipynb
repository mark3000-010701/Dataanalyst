# **[Bài tập] Phân tích dữ liệu Online Retail**

# 1. Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31011319/OnlineRetail.csv',encoding='ISO-8859-1')
data
```

```python
data.info()
data.shape[0]
data.shape[1]
data.columns
data.dtypes
```

# 2. Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data
```

```python
data['Month']=pd.to_datetime(data['InvoiceDate'],format='%Y-%d-%m %H:%M:%S').dt.month
data
```

```python
data['Day']=pd.to_datetime(data['InvoiceDate'],format='%Y-%d-%m %H:%M:%S').dt.day
data['Month']=pd.to_datetime(data['InvoiceDate'],format='%Y-%d-%m %H:%M:%S').dt.month
data['Year']=pd.to_datetime(data['InvoiceDate'],format='%Y-%d-%m %H:%M:%S').dt.year
data['Hour']=pd.to_datetime(data['InvoiceDate'],format='%Y-%d-%m %H:%M:%S').dt.hour
data['Minute']=pd.to_datetime(data['InvoiceDate'],format='%Y-%d-%m %H:%M:%S').dt.minute
data
```

```python
data['InvoiceDate']
```

```python
data.isnull().any()
```

```python
data['Month'].isnull().any()
```

```python
type(data['Month'])
```

```python
conditions = [(data['Month'] >=1) & (data['Month'] <=3),(data['Month'] >=4) & (data['Month'] <=6),(data['Month'] >=7) & (data['Month'] <=9)]
choices = [1,2,3]
data['Previous'] = np.select(conditions,choices,default=4)
data

```

# 3.Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice

```python
data['Amount'] = data['Quantity']*data['UnitPrice']
data
```

# 4. Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4,       5% nếu là ‘France’ ngược lại là 0%.

```python
dk = [(data.Country=='United Kingdom' & (data['Month'] >=10 & (data['Month'] <=12))),(data.Country=='France')]
choices = ['10%','5%']
data['Discount'] = np.select(dk,choices,default='0%')
data
```

# 5. Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.

```python
data['Total'] = data['Amount']-data['Discount']*data['Amount']
data
```
