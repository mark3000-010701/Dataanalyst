# **[Bài tập] Trực quan hóa cơ bản dữ liệu GDP list**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
df=pd.read_csv('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31020936/GDPlist.csv',encoding='ISO-8859-1')
df.head()
```

```python
df.isnull()
```

# So sánh GDP các nước ở South America.

```python
df1=df[df['Continent']=='South America']
df1
```

```python
from matplotlib.pyplot import figure
figure(figsize=(16, 9), dpi=80)
plt.bar(df1['Country'], df1['GDP (millions of US$)'], width = 0.5)
plt.title('GDP in South America', fontsize = 16, color = 'r')
plt.xlabel('Country', fontsize = 14)
plt.ylabel('GDP', fontsize = 14)
plt.show()
```

# **Đánh giá tỉ lệ đóng góp GDP của Việt Nam trên tổng số GDP của 5 nước Đông Nam Á là Vietnam,  Indonesia, Cambodia, Thailand và Malaysia.**

```python
df2=df[df['Continent']=='Asia']
df2
```

```python
df2['Country']=df2['Country'].str.strip()
```

```python
df3=df2[(df2['Country']=='Vietnam')]
df4=df2[(df2['Country']=='Cambodia')]
df5=df2[(df2['Country']=='Malaysia')]
df6=df2[(df2['Country']=='Indonesia')]
df7=df2[(df2['Country']=='Thailand')]
```

```python
df9=df2[(df2['Country']=='Vietnam') | (df2['Country']=='Indonesia') | (df2['Country']=='Cambodia') | (df2['Country']=='Thailand') | (df2['Country']=='Malaysia')]
df9
```

```python
df4
```

```python
from matplotlib.pyplot import figure
figure(figsize=(16, 9), dpi=80)
myexplode = [0, 0, 0, 0,0.2]
plt.pie(df9['GDP (millions of US$)'],labels=df9['Country'],explode=myexplode)
plt.legend(title='GDP of VietNam in South East Asia')
plt.show()
```

# **[Bài tập] Phân tích bộ dữ liệu HousePrice _ Đống Đa**

```python
data=pd.read_excel('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2022/01/08185826/house_price_do%CC%82%CC%81ng-da.xlsx')
data.head()
```

# Vẽ biểu đồ phân tích mối liên hệ giữa diện tích với giá nhà, giữa số phòng ngủ với giá nhà, giữa số toilet với giá nhà.

```python
data1=data[['area','bedroom','toilet','price']]
data1
```

```python
data2=data1[data1.bedroom.isna()==False]
data2
```

```python
data3=data2[data2.toilet.isna()==False]
```

```python
data4=data3[data3.price.isna()==False]
data4
```

```python
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
plt.hist(data4['price'])
plt.show()
```

# Vẽ biểu đồ thể hiện tỉ lệ % bài đăng (bản ghi) giữa các hình thức nhà (type_of_land).

```python
data_type_of_land = data['type_of_land'].value_counts()
data_type_of_land
```

```python
ck=pd.DataFrame(data_type_of_land)
```

```python
from matplotlib.pyplot import figure
figure(figsize=(16, 9), dpi=80)
myexplode = [0.2, 0, 0, 0, 0, 0, 0, 0, 0]
plt.pie(ck)
plt.legend(title='Tỉ lệ bản ghi giữa các hình thức bán nhà( Type_of_land) :')
plt.show()
```

# Vẽ biểu đồ so sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land)

```python
dt = data[['type_of_land','price']]
dt
```

```python
dt=dt[dt.price.isna()==False]
dt
```

```python
dt=dt[dt.type_of_land.isna()==False]
dt
```

```python
dt1 = dt.groupby(['type_of_land'])['price'].mean().reset_index(name='mean_price')
dt1
```

```python
plt.pie(dt1['mean_price'],labels=dt1['type_of_land'])
```

# Vẽ biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng phòng ngủ.

```python
test=data[['bedroom','price']]
test
```

```python
test1=test[test.bedroom.isna()==False]
```

```python
test1=test1[test1.price.isna()==False]
```

```python
check= test1.groupby(['bedroom'])['price'].mean().reset_index(name='mean_for_bedroom')
check
```

```python
plt.bar(check['bedroom'],check['mean_for_bedroom'])
```
