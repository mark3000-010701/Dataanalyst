```python
import pandas as pd
data=pd.read_csv('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31020936/GDPlist.csv',encoding='ISO-8859-1')
data.head()
```

# 2 . Việt hóa tên các cột trong bảng dữ liệu: Country 🡪 Nuoc; Continent 🡪 Chauluc; GDP (millions of US$) 🡪 GDP (trieu $)

```python
data.rename(columns={'Country':'Nuoc','Continent':'Chauluc','GDP (millions of US$)':'GDP(trieu)'},inplace=True)
```

```python
data.head()
```

# 3 . Chèn thêm một cột “Thanhpho” vào sau cột “Nuoc”, giá trị ban đầu là giá trị của cột “Nuoc”

```python
data.head()
```

```python
data.insert(1,"Thanhpho",data.loc[:,'Nuoc'],allow_duplicates=True)
```

```python
data.head()
```

# 4 . Trong cột Thanhpho, thay giá trị Vietnam thành Hanoi; Làm tương tự với các nước còn lại.

```python
data['Thanhpho'] = data['Thanhpho'].replace(regex=r'^Vietna.$',value ="Hanoi")
```

```python
data['Thanhpho'] = data['Thanhpho'].str.strip()
```

```python
data.loc[114:,'Thanhpho']
```

# 5 .Xóa các bản ghi có Chauluc là ‘Asia’

```python
data.drop(data.loc[data.Chauluc=='Asia'].index,inplace=True)
```

```python
data.head(30)
```

# 6 . Xóa các bản ghi có GDP < 300000

```python
data.drop(data.loc[data['GDP(trieu)'] > 300000].index,inplace=True)
```
