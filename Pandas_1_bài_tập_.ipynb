```python
import pandas as pd
data=pd.read_csv('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31020936/GDPlist.csv',encoding='ISO-8859-1')
data.head()
```

# 2 . Việt hóa tên các cột trong bảng dữ liệu: Country 🡪 Nuoc; Continent 🡪 Chauluc; GDP (millions of US$) 🡪 GDP (trieu $)

```python
data.rename(columns={'Country':'Nuoc','Continent':'Chauluc','GDP (millions of US$)':'GDP(trieu)'},inplace=True)
```

```python
data.head()
```

# 3 . Chèn thêm một cột “Thanhpho” vào sau cột “Nuoc”, giá trị ban đầu là giá trị của cột “Nuoc”

```python
data.head()
```

```python
data.insert(1,"Thanhpho",data.loc[:,'Nuoc'],allow_duplicates=True)
```

```python
data.head()
```

# 4 . Trong cột Thanhpho, thay giá trị Vietnam thành Hanoi; Làm tương tự với các nước còn lại.

```python
data['Thanhpho'] = data['Thanhpho'].replace(regex=r'^Vietna.$',value ="Hanoi")
```

```python
data['Thanhpho'] = data['Thanhpho'].str.strip()
```

```python
data.loc[114:,'Thanhpho']
```

# 5 .Xóa các bản ghi có Chauluc là ‘Asia’

```python
data.drop(data.loc[data.Chauluc=='Asia'].index,inplace=True)
```

```python
data.head(30)
```

# 6 . Xóa các bản ghi có GDP < 300000

```python
data.drop(data.loc[data['GDP(trieu)'] > 300000].index,inplace=True)
```

```python
data.head(50)
```

# 1.Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
import pandas as pd
```

```python
df = pd.read_csv('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31011319/OnlineRetail.csv',encoding='ISO-8859-1')
```

```python
df.head()
```

```python
print("số dòng",df.shape[0])
print("số cột",df.shape[1])
```

```python
df.columns
```

```python
df.dtypes
```

# 2.Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv

```python
tg = df.loc[:,['Description','Quantity']]
tg
```

```python
tg.to_csv('OnlineRetail.csv')
```

# 3. Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx

```python
tg1 = df.loc[:999]
tg1
```

```python
tg1.to_excel('OnlineRetail.xlsx')
```

# 4. Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5

```python
tg2 = df[df.Quantity >10]
tg2
```

```python
tg2.to_hdf('OnlineRetail.h5','table')
```

# 5. Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json

```python
tg3 = df.loc[1000:2000,['Quantity','InvoiceDate','UnitPrice']]
tg3
```

```python
tg3.to_json('OnlineRetail.json',orient='columns')
```

# 6. Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html

```python
tg4 = df.loc[df.InvoiceNo == '536365']
tg4
```

```python
tg4.to_html('OnlineRetail.html')
```
