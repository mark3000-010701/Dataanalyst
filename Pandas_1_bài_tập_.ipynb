# 2 . Vi·ªát h√≥a t√™n c√°c c·ªôt trong b·∫£ng d·ªØ li·ªáu: Country ü°™ Nuoc; Continent ü°™ Chauluc; GDP (millions of US$) ü°™ GDP (trieu $)

```python
data.rename(columns={'Country':'Nuoc','Continent':'Chauluc','GDP (millions of US$)':'GDP(trieu)'},inplace=True)
```

```python
data.head()
```

# 3 . Ch√®n th√™m m·ªôt c·ªôt ‚ÄúThanhpho‚Äù v√†o sau c·ªôt ‚ÄúNuoc‚Äù, gi√° tr·ªã ban ƒë·∫ßu l√† gi√° tr·ªã c·ªßa c·ªôt ‚ÄúNuoc‚Äù

```python
data.head()
```

```python
data.insert(1,"Thanhpho",data.loc[:,'Nuoc'],allow_duplicates=True)
```

```python
data.head()
```

# 4 . Trong c·ªôt Thanhpho, thay gi√° tr·ªã Vietnam th√†nh Hanoi; L√†m t∆∞∆°ng t·ª± v·ªõi c√°c n∆∞·ªõc c√≤n l·∫°i.

```python
data['Thanhpho'] = data['Thanhpho'].replace(regex=r'^Vietna.$',value ="Hanoi")
```

```python
data['Thanhpho'] = data['Thanhpho'].str.strip()
```

```python
data.loc[114:,'Thanhpho']
```

# 5 .X√≥a c√°c b·∫£n ghi c√≥ Chauluc l√† ‚ÄòAsia‚Äô

```python
data.drop(data.loc[data.Chauluc=='Asia'].index,inplace=True)
```

```python
data.head(30)
```

# 6 . X√≥a c√°c b·∫£n ghi c√≥ GDP < 300000

```python
data.drop(data.loc[data['GDP(trieu)'] > 300000].index,inplace=True)
```
